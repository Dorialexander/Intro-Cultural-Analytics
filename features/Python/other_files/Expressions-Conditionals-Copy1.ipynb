## Comparison and Conditionals

Last lesson, we talked about the four Python data types, including Booleans or True/False statements. We discussed how if we make a variable called `beyonce` and assign it the value "Grammy award-winner"...

```python
beyonce = "Grammy award-winner"
```

...and then we evaluate that statement with the equals opeartor `==`...

```python
beyonce == "Grammy award-winner"
```

...the operator will return a Boolean. That is to say, it will tell us whether the statement is True or False.

```python
if beyonce == "Grammy award-winner":
    print("Congratulations, Beyonce!")
```

```python
beyonce = "not a Grammy award-winner"
```

```python
if beyonce == "Grammy award-winner":
    print("Congratulations, Beyonce!")
else:
    print("They messed up, Beyonce.")
```

```python
| **Comparison Operator** | **Explanation**                                                                                   |
|:-------------:|:---------------------------------------------------------------------------------------------------:|
| `x == y `         | True if x is equal to y                                                                                |
| `x != y `         | True if x is not equal to y                                               |
| `x > y  `       |  True if is greater than y                                                        |
| `string.endswith('some string')`       |  tests whether string ends with `some string`   
| `string.isspace()`       |  tests whether string is a space |
| `string.replace('old string', 'new string')`      | replaces `old string` with `new string`                                                                             |
| `s.split('delim')`          | returns a list of substrings separated by the given delimiter |
| `s.join(list)`         | opposite of split(), joins the elements in the given list together using the string                                                                        |
                                                            
```
