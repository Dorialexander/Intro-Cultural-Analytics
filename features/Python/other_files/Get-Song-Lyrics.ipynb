```python
def get_all_songs_from_album(artist, album_name):
    artist = artist.replace(" ", "-")
    album_name = album_name.replace(" ", "-")
    resp = requests.get(f"https://genius.com/albums/{artist}/{album_name}")
    html_str = resp.text
    document = BeautifulSoup(html_str, "html.parser")
    title_tags = document.find_all("h3", attrs={"class": "chart_row-content-title"})
    song_titles = [title.text for title in title_tags]
    clean_songs = []
    for song in song_titles:
    #re.split(r'\W', song)
    #song.rstrip()
        #clean_song = song.replace(" ", "")
        #clean_song = re.search('(?<=\\n).*?(?=\\n)', clean_song).group(0)
        clean_song = re.match('^[^\(]+', song).group(0)
        clean_song = re.sub('Lyrics', "", clean_song).strip()
        clean_song = clean_song.strip()
        clean_song = unicodedata.normalize("NFKD", clean_song)
        clean_songs.append(clean_song)
    return clean_songs
```

```python
def download_album_lyrics(artist, album_name): 
    clean_songs = get_all_songs_from_album(artist, album_name)
    genius = lyricsgenius.Genius(client_access_token)
    genius.remove_section_headers = True
    hyphen_artist = artist.replace(" ", "-")
    
    album_title = album_name.replace(" ", "-")
    album_lyrics = []
    for clean_song in clean_songs:
        song_title = clean_song
        song_title = song_title.replace(" ", "-")
        
        song_object = genius.search_song(clean_song, artist)

        filename = f'{hyphen_artist}-{album_title}/{song_title}.txt'
        filename = re.sub('(-){2,}', '', filename)

        os.makedirs(os.path.dirname(filename), exist_ok=True)
        
        if song_object != None:
            lyrics = song_object.lyrics
            album_lyrics.append(lyrics)
        
            with open(filename, 'w') as file_object:
                file_object.write(lyrics)
        else:
            print('No lyrics')
    with open(f'{hyphen_artist}-{album_title}/{artist}-{album_title}-Album-Lyrics.txt', 'w') as album_file_object:
        for song in album_lyrics:
            album_file_object.write(song)
        #album_lyrics.append(song_lyrics)
        #song_lyrics.save_lyrics(extension='txt', filename=f'{album_title}/{song_title}', sanitize=True)
    print(f'✨Download of {artist}\'s album {album_title} complete.✨')
```
