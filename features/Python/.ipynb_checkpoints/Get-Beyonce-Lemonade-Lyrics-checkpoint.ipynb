## Genius Song Lyric API

```python
from secrets import *
```

```python
import lyricsgenius
```

```python
genius = lyricsgenius.Genius(client_access_token)
```

```python
genius.remove_section_headers = True
```

```python
def get_all_songs_from_album(artist, album_name):
    resp = requests.get(f"https://genius.com/albums/{artist}/{album_name}")
    html_str = resp.text
    document = BeautifulSoup(html_str, "html.parser")
    title_tags = document.find_all("h3", attrs={"class": "chart_row-content-title"})
    song_titles = [title.text for title in title_tags]
    clean_songs = []
    for song in song_titles:
    #re.split(r'\W', song)
    #song.rstrip()
    #song.replace(" ", "")
        clean_song = re.search('(?<=\\n).*?(?=\\n)', song).group(0)
        clean_song = clean_song.strip()
        clean_song = unicodedata.normalize("NFKD", clean_song)
        clean_songs.append(clean_song)
    return clean_songs
```

```python
get_all_songs_from_album('Beyonce', 'Lemonade')
```

```python
album_title = 'Lemonade/'
for song in clean_songs:
    song_lyrics = genius.search_song(song, 'Beyonce')
    song_title = song.replace(" ", "-")
    #print(song.lyrics)
    song_lyrics.save_lyrics(extension='txt', filename=f'{song_title}', sanitize=False)
```
