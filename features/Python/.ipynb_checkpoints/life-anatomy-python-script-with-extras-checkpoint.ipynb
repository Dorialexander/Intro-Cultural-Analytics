## The Life and Anatomy of a Python Script

A lot of Python tutorials begin with the nitty-gritty, but we're going to start from the abstract, from a zoomed-out perspective. We're going to take a look at the life and anatomy of a Python script.

## Import libraries and modules

```python
import pandas as pd
import csv
import spacy
```

```python
from collections import Counter
import re
from spacy.lang.en.stop_words import STOP_WORDS
```

## Define functions

```python
def split_into_words(full_text):
    words = re.split(r"\W+", full_text.lower())
    return words 
```
