```python
def split_text_by_delimiters(input_text_file, output_text_file, first_split_phrase='About the Publisher', last_split_phrase="ACKNOWLEDGMENTS"):
    text = open(input_text_file).read()
    text_sans_intro = text.split(first_split_phrase)[1].strip()
    desired_text = text_sans_intro.split(last_split_phrase)[0].strip()
    
    output_file = open(f'{output_text_file}', "w")
    output_file.write(desired_text)
```

```python
def shuffle_lines_per_page(page, n=5):
    
    all_pages = []
    """Yield successive n-sized chunks from lst."""
    for i in range(0, len(page), n):
        
        page_lines = page[i:i + n]
        random.shuffle(page_lines)
        all_pages.append(page_lines)
    return all_pages
```
