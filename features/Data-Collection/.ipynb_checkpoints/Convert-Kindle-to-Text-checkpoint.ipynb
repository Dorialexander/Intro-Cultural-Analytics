## Convert Kindle Book to Plain Text File

## Purchase and Download Kindle Book

- Purchase your Kindle book
- Go to your Amazon account -> Manage content and devices

<img src="../images/Calibre/Manage-content-devices.png" >

- Click on the ellipsis "..." next to the text that you want to download and select "Download & transfer via USB"

<img src="../images/Calibre/Download-USB-Dreams.png" >

The file should download to your computer.

## Download and Install Calibre

Download Calibre: https://calibre-ebook.com/download

## Download and Install DeDRM Plugin

Download the latest release of DeDRM tools by clicking "DeDRM_tools_6.7.0.zip" (or whatever version number is most recent) here: https://github.com/apprenticeharper/DeDRM_tools/releases

Unzip the file but do not unzip the plugins inside this folder.

## Load DeDRM Plugin

Go to Calibre -> Preferences -> Plugins -> Load plugin from file. Then navigate to the DeDRM_Plugin.zip file inside DeDRM_tools_6.7.0 folder. 

<img src="../images/Calibre/Preferences.png" >

<img src="../images/Calibre/Plugins.png" >

## Add Kindle Serial Number

Finally, add your Kindle serial number to the plugin. To find your Kindle serial number, go to your Amazon account -> Manage content and devices -> Devices

<img src="../images/Calibre/Serial-Number.png" >

Then go back to Calibre's Preferences -> Plugins. Select "Show only user installed plugins", click on the DeDRM plugin and select "Customize plugin". Select "eInk Kindle ebooks" and add your Kindle serial number.

<img src="../images/Calibre/Customize-eInk.png" >

## Convert Book to Text File

Now drag and drop your Kindle book (.azw) into Calibre. Once it's loaded, select "Convert books" form the Calibre menu, which will cause a new window to appear. Then choose "TXT" as the desired output format and select "OK".

<img src="../images/Calibre/Output-Format.png" >

You should find the plain text file in a directory called "Calibre Library", which should be in your home folder.

<img src="../images/Calibre/Where-to-Find-Text-File.png" >
