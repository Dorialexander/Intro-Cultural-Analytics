## How To Keep Twarc Running

`nohup twarc filter "search term" > search_term.jsonl &`

## How To Stop Twarc From Running

Find the process ID and kill the process by process ID

 Mac/Chrome OS

```python
!ps aux | grep twarc
```

```python
!kill 80884
```

Kill the process by name

 Mac/Chrome OS

```python
!pkill -f "twarc"
```
